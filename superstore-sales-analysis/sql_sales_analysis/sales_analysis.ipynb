## Project: Sales Analysis using Python+SQL

```python
# Download the dataset
import kaggle
kaggle.api.authenticate()
```

```python
kaggle.api.dataset_download_files(dataset='bhanupratapbiswas/superstore-sales', path='data', unzip=True)
```

```python
# Load data into pandas
import pandas as pd
df = pd.read_csv("./data/sales.csv", encoding='latin-1')
```

```python
# view
df.head()
```

```python
# Rename the columns
df.rename(columns = {col:col.lower() for col in df.columns}, inplace=True)
```

```python
# Check for duplicates
df.duplicated().sum()
```

```python
df = df.drop(columns=['row_id'])
```

```python
df.head()
```

```python
# Change the data type of date columns from object to datetime format
df['order_date'] = pd.to_datetime(df['order_date'], format='%d/%m/%Y').dt.date
df['ship_date'] = pd.to_datetime(df['ship_date'], format='%d/%m/%Y').dt.date
```

```python
# view
df.head()
```

```python
# Create a MySQL connection
import sqlalchemy as sal
engine = sal.create_engine("mysql://root:1234@localhost/mydb")
conn = engine.connect()
```

```python
# load data into sql server
df.to_sql('df_order', con=conn, index=False, if_exists='append')
```

```python
conn.close()
```
